# The algorithm

We start with the problem
$$\frac{100}{7}$$

I have a guess for what the answer is. The guess is 10. In other words I guess that 7 times 10 is 100.

This is clearly wrong, but I can now formulate the problem as finding the error between my guess and the actual answer.

Okay so we have that $7*10 = 70$

I want to shift the guess of 10 by some value to the real answer.

Okay so we have that $7*(10-s) = 70$ where s is the shift to the correct answer

However, I also need to shift 70 somehow to get to 100.

Okay so we have that $7*(10-s) = 70-s$ where s is the shift to the correct answer

However, I know that the shift s may not change 10 and 70 at the same rate, so I'll add one more term w that changes s into the correct shift

Okay so we have that $7*(10-s) = 70-sw$ and then I can relate the right side to $70-sw = 100$

First I'll reduce down first equation.

$7*(10-s) = 70-sw$ is really just $70 - 7s = 70-sw$ which means that w must be 7. Note that 7 was the denominator in the first division equation.

anyway we have that $w=7$ so the right side is now $70-7s$ which of course equaled 100. 

Now it's a simple solve as $-7s = 100-70$ which is just $s = -\frac{30}{7}$.

Okay so now I know the $guess - s$ is the correct solution to give me 100. But now I don't know what $-\frac{30}{7}$ equates to. So I keep repeating until I get to a fraction easily solvable.

So I do original guess - shift where shift is $(numerator - denominator*guess) / -denominator$ or  $(n - d*g) / -d$ 

If I can automatically make guesses, then I can automate the recursive algorithm


```python
def num_digits(n):
	return len(str(n))
def guess(n, d):
	diff = num_digits(n) - num_digits(d)
	return d*10**(diff - 1)

def cracked_div(n, d):
	# base case should be a codebook of known divisions that are reasonable easy
	g = guess(n, d)
	shift_n = -(n - d*g)
	shift_d = d
	print(shift_n/shift_d)
	return g - cracked_div(shift_n, shift_d)

cracked_div(100, 7)
```
